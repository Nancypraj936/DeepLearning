```python
#Test The accuracy of MNIST dataset using Batch Normalization with Conv2D
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization, ReLU,Conv2D,MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

```python
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Preprocess the data
x_train = x_train.reshape((60000, 28, 28, 1)) / 255.0 # Reshape and normalize training images
x_test = x_test.reshape((10000, 28, 28, 1)) / 255.0 # Reshape and normalize test images
y_train = to_categorical(y_train) # Convert training labels to categorical format
y_test = to_categorical(y_test) # Convert test labels to categorical format
```

```python
# Define the model architecture
model = Sequential([ Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
BatchNormalization(),
Conv2D(64, (3, 3), activation='relu'),
BatchNormalization(),
MaxPooling2D((2, 2)),
Flatten(),
Dense(128, activation='relu'),
BatchNormalization(),
Dense(10, activation='softmax')])
```

```python
# Compile the model
model.compile(optimizer='adam',
loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train, y_train, batch_size=128, epochs=5,
validation_data=(x_test, y_test))
```

```python
# Evaluate Model
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {acc * 100:.2f}%")
```
