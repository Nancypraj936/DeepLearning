```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
#Load breast cancer dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
# keys in dataset
cancer_dataset_keys=cancer_dataset.keys()
#print("cancer_dataset Keys:",cancer_dataset_keys)
# name of features
cancer_dataset_feature_names=cancer_dataset['feature_names']
#print("cancer_dataset_feature_names::",cancer_dataset_feature_names)
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
columns = np.append(cancer_dataset['feature_names'],['target']))
#print("Cancer Data Frame::",cancer_df)
```

```python
# cancer_df.head(6)
# cancer_df.info()
# cancer_df.describe()
```

```python
# input variable
X = cancer_df.drop(['target'], axis = 1)
# output variable
y = cancer_df['target']
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size = 0.2, random_state= 5)
```

```python
# Support vector classifier
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
y_pred_scv = svc_classifier.predict(X_test)
accuracy=accuracy_score(y_test, y_pred_scv)
print("Accuracy Score::",accuracy)
```
