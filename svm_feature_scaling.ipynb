```python
#Load the breast cancer dataset & predict the data using SVM algorithm with feature scaling
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
#Load breast cancer dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
# keys in dataset
cancer_dataset_keys=cancer_dataset.keys()
#print("cancer_dataset Keys:",cancer_dataset_keys)
```

```python
# name of features
cancer_dataset_feature_names=cancer_dataset['feature_names']
#print("cancer_dataset_feature_names::",cancer_dataset_feature_names)
# create datafrmae
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
columns = np.append(cancer_dataset['feature_names'], ['target']))
#print("Cancer Data Frame::",cancer_df)
```

```python
# cancer_df.head(6)
# cancer_df.info()
# cancer_df.describe()
```

```python
# input variable
X = cancer_df.drop(['target'], axis = 1)
# output variable
y = cancer_df['target']
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size = 0.2, random_state= 5)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
# Support vector classifier
# Train with Standard scaled Data
from sklearn.svm import SVC
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc_classifier2.predict(X_test_sc)
score=accuracy_score(y_test, y_pred_svc_sc)
print("Accuracy Score:",score)
```
