```python
#Plotting the breast cancer dataset using pairplot
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load breast cancer dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
# keys in dataset
cancer_dataset_keys=cancer_dataset.keys()
print("cancer_dataset Keys:",cancer_dataset_keys)
```

```python
# name of features
cancer_dataset_feature_names=cancer_dataset['feature_names']
print("cancer_dataset_feature_names::",cancer_dataset_feature_names)
```

```python
# create datafrmae
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],
cancer_dataset['target']],
columns = np.append(cancer_dataset['feature_names'], ['target']))
print("Cancer Data Frame::",cancer_df)
```

```python
# pair plot of sample feature
sns.pairplot(cancer_df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'] )
plt.show()
```
